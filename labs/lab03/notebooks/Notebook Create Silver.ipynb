# Create Silver

## Reset

```sql
DROP TABLE IF EXISTS dim_school;
DROP TABLE IF EXISTS dim_schoolType;
DROP TABLE IF EXISTS dim_year;
DROP TABLE IF EXISTS fact_attendance;
```

## Main

### DDL

```sql
CREATE TABLE IF NOT EXISTS dim_school (
    dsId                INT NOT NULL,
    deReferenceNumber   INT NOT NULL,
    schoolName          STRING NOT NULL,
    country             STRING NOT NULL
    );
```

```sql
CREATE TABLE IF NOT EXISTS dim_schoolType (
    stId                INT NOT NULL,
    schoolType          STRING NOT NULL
    );
```

```sql
CREATE TABLE IF NOT EXISTS dim_year (
    yearId              INT NOT NULL,
    yearDesc            STRING NOT NULL
    );
```

```sql
CREATE TABLE IF NOT EXISTS fact_attendance (
    faId                INT NOT NULL,
    dsId                INT NOT NULL,
    yearId              INT NOT NULL,
    schoolName          STRING NOT NULL,
    percentAttendance   DECIMAL(5,2) NOT NULL
    );
```
